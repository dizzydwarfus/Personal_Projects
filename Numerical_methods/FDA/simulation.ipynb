```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Define boundary conditions and constants

```python
Nt = 1000 # number of temporal steps
Nx = 100 # number of spatial steps

tMin = 0.1 # in seconds
tMax = 10 # in seconds

xMin = 0 # in meters
xMax = 10 # in meters
```

```python
t = np.linspace(tMin, tMax, Nt)
x = np.linspace(xMin, xMax, Nx)
```

#### Compute secondary parameter

```python
deltaT = t[1] - t[0]
deltaX = x[1] - x[0]
```

#### Numerical solve the PDE

##### Initialize a container to hold all the p value

```python
p = np.zeros((Nt,Nx))
p.ndim, p.shape, p.size
```

##### Set initial conditions

```python
# define gaussian function

def analytical_sol(time, space):
    sol = (1/np.sqrt((2*np.pi*time)))*np.exp((-space**2)/(2*time))
    return sol
```

```python
p[0,:] = analytical_sol(t[0], x)
```

```python
plt.plot(x, p[0,:])
```
